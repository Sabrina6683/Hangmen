```python
hangman={0:'''
        ____________
         |''',
        1:'''
        ____________
         |
         O''',
        2:'''
        ____________
         |
         O
        /''',
        3:'''
        ____________
         |
         O
        / \\''',
        4:'''
        ____________
         |
         O
        / \\
         |''',
        5:'''
        ____________
         |
         O
        / \\
         |
        /''',
        6:'''
        ____________
         |
         O
        / \\
         |
        / \\ '''}

type(hangman)
```

```python
from random import randrange                                                            # Import module random to use randrage() method to pick random words from a defined list
list_of_words = ["FOREST", "SCHOOL", "LEARNING", "CAMPING", "APPLETREE"]                # words to pick from
word = list_of_words[randrange(len(list_of_words))]                                     # defines variable that picks randomly from lists of words
hidden_word = len(word)*"-"                                                             # Create a hidden word with the same length as the secret word
print(f"Hidden word is {hidden_word}")

mistake_count = 6
thought_ballon = "\U0001F4AD" * 15                                                      # Create a variable that defines the look of a divider
print(thought_ballon)
while mistake_count > 0 and hidden_word.count("-") != 0:                                # while condition specifies conditions that have to be fullfilled
  letter_count = 0                                                                      # new variable necessary to change hidden word if a correct letter was choosen
  index_count = 0                                                                       # new variable necessary to change right dash with correct letter
  print("Pick a letter and press ENTER")
  picked_letter = input().capitalize()                                                  # Create an input variable, that will output capitalized letters
  for letter in word:                                                                   # for condition creates temp. variable letter that views at individual letters in word
    if letter == picked_letter:                                                         # if condition creates new condition: if letter in the word is equal to picked letter execute the following actions
      hidden_word = hidden_word[:index_count] + picked_letter + hidden_word[index_count+1:]     # action to previous if condition: change hidden word by changing dash in correct position to the correct letter
      letter_count += 1
    index_count += 1
  if word.count(picked_letter) > 0:                                                     # if the number of the selected letter in the word is greater than 0, then execute the following actions
    print(f"CORRECT! =) The word contains the {picked_letter}")
    print(hidden_word)
    print(thought_ballon)
  else:                                                                                 # if not execute the following action (reduce the mistake_count and print...)
    mistake_count -= 1
    print(f"WRONG! =( Number of mistakes left: {mistake_count}")
    print(hangman[6 - mistake_count])
    print(thought_ballon)
if mistake_count == 0:                                                                  # if amount of choosen letters is equal to 0 all attempts are used up
    print("HANGED!!! =((( The word you were looking for was:", word)
else:                                                                                   # if conditions from while condition are not fullfilled do next action
  print("CONGRATULATIONS! =)))")
```
